# Embeddings with Python Gemini API

**Embeddings** ka simple matlab hai ke text ko numbers ke form mein convert karna, taake computer usay asani se samajh sake aur compare kar sake. Numbers ki list ko "array" kehte hain, aur isi array ko embedding kehte hain.

Yeh embeddings kaam aati hain jab aapko texts ke darmiyan similarities ya differences dhoondhne hoon, jaise:

- Dono texts ek hi topic par hain ya nahi.
- Dono texts ka sentiment (positive ya negative) ek jaisa hai ya nahi.

Embeddings ko generate karne ke liye `embed_content` method use hota hai. Yeh method alag-alag tasks ke liye kaam karta hai, jo niche explain kiye gaye hain:

## Task Types aur Unka Matlab:

### **RETRIEVAL_QUERY**
- Iska matlab yeh hai ke aapka text ek search query hai (jaise Google par search karte hain).
- **Example:**
  ```
  "Best laptops under $1000"
  ```

### **RETRIEVAL_DOCUMENT**
- Isme aap ek document ka text aur uska title dete hain.
- Yeh retrieval systems (search engines) mein kaam aata hai.
- **Example:**
  - **Title:** "Top Laptop Brands"
  - **Text:** "This document lists the top laptop brands of 2025."

### **SEMANTIC_SIMILARITY**
- Iska matlab yeh hai ke aap do texts ke darmiyan similarity check karna chahte hain.
- **Example:**
  ```
  "Cats are cute pets" aur "Cats make great companions"
  ```

### **CLASSIFICATION**
- Iska use text ko kisi category mein dalne ke liye hota hai.
- **Example:**
  - Ek review ka text classify karna (positive ya negative).

### **CLUSTERING**
- Iska matlab yeh hai ke similar texts ko groups (clusters) mein organize karna.
- **Example:**
  - Blog articles ko topics ke mutabiq group karna.

## Simple Steps:
1. Gemini API ka Python SDK install karo.

3. Apne task ka type specify karo (e.g., `RETRIEVAL_QUERY` ya `CLASSIFICATION`).
4. Result mein aapko numbers ki list (embedding) milegi, jo comparison ke liye use hoti hai.

# **Retreivers**

````python
from langchain_core.documents import Document
from langchain_core.runnables import RunnableLambda

retriever = RunnableLambda(vectorstore.similarity_search).bind(k=1) # select top result
retriever.batch(["shark"])
````

### **1. `bind()`**

**`bind()` ka kaam:**  
- Kisi function ke **default arguments fix (yani bind)** kar deta hai, taake baar-baar unhe specify na karna pade.

#### **Example Without `bind`**:
```python
result = vectorstore.similarity_search("cat", k=1)  # Har baar `k=1` specify karna padega
result2 = vectorstore.similarity_search("dog", k=1)
```

#### **Example With `bind`**:
```python
retriever = RunnableLambda(vectorstore.similarity_search).bind(k=1)
result = retriever("cat")  # Ab `k=1` automatic apply ho gaya
result2 = retriever("dog")  # Fir se `k=1` automatic apply
```

**Benefit of `bind`:**  
- Reusable bana deta hai, aur default arguments (yahan `k=1`) ko fix kar deta hai. Aapko har baar `k=1` dene ki zarurat nahi hoti.

---

### **2. `batch()`**

**`batch()` ka kaam:**  
- Ek **saath multiple queries** run karta hai. Agar ek query deni ho, to aap directly function call karte hain, lekin agar multiple queries deni hoon, to `batch()` ka use hota hai.

#### **Example Without `batch`**:
```python
result1 = retriever("cat")  # Sirf ek query chalai
result2 = retriever("dog")  # Doosri query chalai
```

#### **Example With `batch`**:
```python
results = retriever.batch(["cat", "dog"])  # Ek saath dono queries chalai
```

**Output:**  
- Aapko ek **list of results** milegi:  
  ```python
  [
      Result for "cat",
      Result for "dog"
  ]
  ```

````python
# Preprocessing function to transform the image into a tensor

def preprocess_image(image_path):
  image = Image.open(image_path).convert('RGB')
  preprocess = transforms.Compose([
  transforms.Resize((224, 224)),
  transforms.ToTensor(),
  transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
  ])
  return preprocess(image).unsqueeze(0)

# Function to create image embeddings
def create_image_embedding (image_path):
  try:
      input_tensor = preprocess_image(image_path)
      with torch.no_grad():
         embeddings = model(input_tensor)# embedding important line
      return embeddings.squeeze().numpy()
  except Exception as e:
    print("Error:", e)
    return None
````

````python
# prompt: create python function where we provide image url and imag_name then the function will save it image in image folder

import requests
import os

def save_image_from_url(image_url, image_name):
  """
  Downloads an image from a URL and saves it to the 'images' folder.

  Args:
    image_url: The URL of the image to download.
    image_name: The name of the file to save the image as.
  """

  try:
    if not os.path.exists("images"):
      os.makedirs("images")

    image_path = os.path.join("images", image_name)

    response = requests.get(image_url, stream=True)
    response.raise_for_status() # Raise an exception for bad status codes

    with open(image_path, 'wb') as file:
      for chunk in response.iter_content(chunk_size=8192):
        file.write(chunk)

    print(f"Image saved to: {image_path}")
  except requests.exceptions.RequestException as e:
    print(f"Error downloading image: {e}")
  except Exception as e:
    print(f"An error occurred: {e}")
````
